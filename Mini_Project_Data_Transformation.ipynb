```python
import requests as rq
import pandas as pd
import numpy as np
from io import BytesIO
import xlrd
from datetime import datetime
from dateutil import parser
```

```python
fileName = 'Food+Orders.xlsx'
filePath = 'https://raw.githubusercontent.com/shashubh/bits_aiml_python_pub/main/'
```

```python
url = filePath+fileName
data = rq.get(url).content
df = pd.read_excel(BytesIO(data))
df = pd.DataFrame(df)
```

```python
df.info()
```

# **Data Transformation**

1. Split the Items column based on semicolon(;) into different columns and include the splitted data. 
2. remove/drop the column from the data frame. 

```python
newItemsCols = df.Items.str.split(';', expand=True)
for i in range(0, newItemsCols.shape[1]):
  df['Items_'+str(i+1)] = newItemsCols[[i]]
df.drop(['Items'], axis=1)
```

2. Handle the Date/Time columns for Date (Booking Date) and Delivery Time.   

```python
df['DateTime'] = pd.to_datetime(df['DateTime'], format='%d-%B-%Y %I.%M %p')
df['Delivery Time'] = pd.to_datetime(df['Delivery Time'], format='%d-%B-%Y %I.%M %p')
df['ActualDeliveryDuration'] = df['Delivery Time'] - df['DateTime']
df.drop(['DateTime','Delivery Time'], axis=1)
```

```python
df['FinalCost'] = df['Cost'] - df['Discount'] + df['Delivery charges'] + df['Surge charges'] + df['Packaging charges'] +df['ST'] + df['Tip']
df.drop(['Cost', 'Discount', 'Delivery charges', 'Surge charges', 'Packaging charges', 'ST', 'Tip'], axis=1)
```
